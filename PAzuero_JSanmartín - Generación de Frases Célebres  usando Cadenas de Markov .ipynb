En este cuaderno se presenta un pequeño ejemplo de como se generan palabras usando cadenas de marcov, para ello nos hemos basado en el cuaderno facilitado por nuestro docente. En el presente ejemplo tomaremos de referencia a una página web de frases célebres.

### 1. Obtencion de  los datos ###

```python
from bs4 import BeautifulSoup
import requests

print('Módulos importados')

#1. Obteniendo datos
r = requests.get("https://www.psicoactiva.com/blog/frases-celebres/")
soup = BeautifulSoup(r.text)
#print(soup)
frases = map(lambda x: x.text.replace("\"",""),soup.select("p"))
#frases = map(lambda x: x.replace("\n",""),frases)
frases = map(lambda x: x.replace("-",""),frases)
frases = map(lambda x: x.replace("\t",""),frases)
#frases = map(lambda x: x.replace(" ",""),frases)
palabras = map(lambda x: str(x).split(" "),frases)
print(palabras)

#listaPalabras=list(palabras)
#print(listaPalabras)
```

### 2. Generación del grafo de Markov ###

```python
#2. Generando el grafo de Markov
from collections import defaultdict
almacen = defaultdict(lambda: list())

def add_word(prev,next):
    global almacen
    almacen[prev].append(next)

for frase in palabras:
    frase = list(frase)
    #print(frase)
    for i,palabra in enumerate(frase):
        if i == 0:
            add_word(("START","START"),palabra)
            continue
        if i == 1:
            add_word(("START",frase[0]),palabra)
            continue
        add_word((frase[i-2],frase[i-1]),palabra)
        
print(almacen)
#listaAlmacen=list(almacen)
#print(listaAlmacen)
```

### 3.Generando de frase nueva ###

```python
#3. Generando una frase nueva
import random

#Función para generar nueva frase
def gen_word():
    word = list()
    state = "START","START"
    while True:
        w = random.choice(almacen[state])
        word.append(w)
        state = state[1],w
        if w.endswith(".") or w.endswith("!") or w.endswith("\n"):
            return " ".join(word)
```

### 4. Presentar los resultados ###

```python
#4. Generando frases

#Papá, papá los mosquitos me quieren picar!
#Papá, papá los mosquitos me vienen a picar con linternas!
for a in range(40):
    try:
        palabraGenerada=gen_word()
        print(a,' ',palabraGenerada)
    except Exception as err:
        continue
```
